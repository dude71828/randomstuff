```python
import time
import numpy as np
import random
import matplotlib.pyplot as plt

def detm(size):
    a = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            a[i, j] = random.random()
    return np.linalg.det(a)
 
def timing(size):
    start = time.perf_counter()
    _ = detm(size)
    end = time.perf_counter()
    return (end - start)

times = [] 
max = 200
reps = 30
for i in range(1, max):
    size = i
    t = 0
    for i in range(reps):
        t += timing(size)
    times.append(t / reps)

plt.plot(range(1, max), times)
plt.xlabel('Matrix Size')
plt.ylabel('Time (seconds)')
plt.title('Determinant Calculation Time vs. Matrix Size')
plt.grid()
```

```python
import time
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats


def detm(size):
    a = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            a[i, j] = 10*random.random()
    return size*np.linalg.det(a)/np.sum(a)

data = []
for i in range(1000):
    data.append(detm(13))


sns.kdeplot(data)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('KDE Plot of Data')
plt.show()

print("Count:", len(data))
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Min:", np.min(data))
print("Max:", np.max(data))
print("25th percentile (Q1):", np.percentile(data, 25))
print("75th percentile (Q3):", np.percentile(data, 75))

stats.probplot(data, dist="norm", plot=plt)
plt.title("Normal Probability Plot (Q-Q Plot)")
plt.show()
```
